# Importando bibliotecas necessárias

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

# Gerando dados simulados: suponha que esses são registros de log com duas características (por exemplo, frequência e tipo de evento)

# Neste exemplo, estamos gerando 300 pontos de dados com 4 centros (clusters) distintos

```python
dados, verdadeiros_centros = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

# Visualizando os dados gerados: imporante para entender a distribuição dos dados antes da aplicação do K-means

```python
plt.scatter(dados[:,0],dados[:,1], s=50)
plt.title("Dados Simulados de Eventos de Segurança")
plt.xlabel("Característica 1 (ex: frequência)")
plt.ylabel("Característica 2 (ex: tipo)")
plt.show()
```

# Aplicando o algoritmo K-means

# Neste exemplo, definimos 4 clusters, que é o número de centros que usamos para gerar os dados

```python
kmeans = KMeans(n_clusters=4, n_init=2)
kmeans.fit(dados)
```

# Obtendo os centros dos clusters e as previsões de qual cluster cada ponto de dados pertence

```python
centros = kmeans.cluster_centers_
previsoes = kmeans.predict(dados)
```

# Visualizando os clusters formados e seus centros

```python
plt.scatter(dados[:, 0], dados[:, 1], c=previsoes, s=50, cmap='viridis')
plt.scatter(centros[:, 0], centros[:, 1], c='red', s=200, alpha=0.5)
plt.title("Clusters Identificados pelo K-Means")
plt.xlabel("Característica 1 (ex: frequência)")
plt.ylabel("Característica 2 (ex: tipo)")
plt.show
```
