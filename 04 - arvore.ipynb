# Importações

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

# Simulando um conjunto de dados de e-mails

```python
dados_email = pd.DataFrame({
    'palavras_suspeitas': [1,0,0,1,1,0], #1 para sim, 0 para não
    'dominio_confiavel': [0,1,1,0,0,1], #1 para sim, 0 para não
    'classificacao': [1,0,0,1,1,0] #1 para phishing, 0 para seguro
})
```

# Separando as características e o alvo

```python
x = dados_email[['palavras_suspeitas', 'dominio_confiavel']]
y = dados_email['classificacao']
```

# Criando o modelo de Árvre de Decisão

```python
modelo = DecisionTreeClassifier()
modelo.fit(x, y)
```

# Visualizando a Árvore de Decisão

```python
plt.figure(figsize=(10,8))
tree.plot_tree(modelo, filled=True, rounded=True,
                feature_names=['palavras_suspeitas', 'dominio_confiavel'],
                class_names=['Seguro', 'Phishing'])
plt.show()
```
